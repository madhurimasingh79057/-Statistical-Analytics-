## MEAN

```python
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]   # List of numbers
mean = sum(x) / len(x)            # Calculating mean: sum of numbers / total count
print(mean)                       # Printing the mean value
```

## CALCULATING MEAN

```python
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]   # List of numbers
mean = sum(x) / len(x)            # Calculating mean: sum of numbers / total count
print(mean)                       # Printing the mean value
```

```python
population=[150,160,170,165,155]
pop_mean=sum(population)/len(population)
print("population mean is:",pop_mean)
```

```python
sample=[150,170,165]
sam_mean=sum(sample)/len(sample)
print("sample mean is:",sam_mean)
```

## USING NUMPY LIBRARY

```python
import numpy as np
data=[10,20,30,40,50]
sample_mean=np.mean(data)
print("mean is the:",sample_mean)
```

## USING STATISTICS

```python
import statistics as st
data=[20,40,60,80,100]
mean_st=st.mean(data)
print("mean is the:",mean_st)
```

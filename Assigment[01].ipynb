```python
data = [5,8,None,7,10,None,6]
s = 0
count = 0
for x in data:
 if x is not None:
    s+=x
    count+=1
mean_va1 = s/count
imputed = []
for x in data:
  if x is None:
    imputed.append(mean_va1)
  else:
    imputed.append(x)
  print("original data:",data)
  print("imputed data:",imputed)
```
